```python
def GetPosFromQueryResult(QS):
    pos = []
    for j in QS.raw_response['results']:
        for i in j:
            if(i == "geometry"):
                pos.append({"lat":  float(j[i]['location']['lat']) , "lon" : float(j[i]['location']['lng'])  } )
    print("the num of data: ", len(pos))
    return pos
```
